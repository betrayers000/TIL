```python
ans = []
def genius(N,K,key,**keywargs):
    for i in range(key,K+1):
        if N != 0:
            keywargs[N]=i
            key += 1
            genius(N-1,K,key,**keywargs)
        else:
            global ans
            ans.append(keywargs)
dic = {}
genius(3,5,1,**dic)
print(ans)
```
